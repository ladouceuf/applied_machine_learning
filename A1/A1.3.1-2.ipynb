```python
def find_w_star(x,y):
    xtx=x.T.dot(x) 
    xtx_inv=np.linalg.inv(xtx)
    w_star=xtx_inv.dot(x.T.dot(y))
    return w_star

def find_MSE(x,y,w):
    return (y-x.dot(w)).T.dot((y-x.dot(w)))/x.shape[0]
```

```python
MSE_train=[]
MSE_test=[]
w_star_lst=[]

#Loop though the training files
for i in range(1,6):
    path="out/3.2/CandC/CandC-train"
    path=path+str(i)+".csv"
    df=pd.read_csv(path)
    x=df.iloc[:,0:123]
    x=np.array(x)
    y=df.iloc[:,123:]
    y=np.array(y)
    #Compute minimal loss solution w_star
    w_star=find_w_star(x,y)
    w_star_lst.append(w_star)
    #Compute minimal MSE
    MSE=find_MSE(x,y,w_star)
    #push value into the MSE_train list
    MSE_train.append(MSE)
    
#Loop though the test files
for i in range(1,6):
    path="out/3.2/CandC/CandC-test"
    path=path+str(i)+".csv"
    df=pd.read_csv(path)
    x=df.iloc[:,0:123]
    x=np.array(x)
    y=df.iloc[:,123:]
    y=np.array(y)
    #Compute minimal MSE
    MSE=find_MSE(x,y,w_star)
    #push value into the MSE_test list
    MSE_test.append(MSE)

#Average the MSE
MSE_train_av=sum(MSE_train)/float(len(MSE_train))
MSE_test_av=sum(MSE_test)/float(len(MSE_test))
```

```python
#Convert lists into np.array's
MSE_train=np.array(MSE_train)
MSE_test=np.array(MSE_test)
#Print the results
print "These are the MSE of the training datasets for the 5 different splits: \n", MSE_train,"\n"
print "This is the 5-fold cross-validation error for the training dataset: ", MSE_train_av,"\n"
print "These are the MSE of the testing datasets for the 5 different splits: \n", MSE_test,"\n"
print "This is the 5-fold cross-validation error for the testing dataset: ", MSE_test_av,"\n"
```
